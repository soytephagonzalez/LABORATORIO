# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
```

```python
# Ejecutar la consulta SQL para obtener los datos de la columna ConvertedComp
QUERY = """
SELECT ConvertedComp
FROM master
"""

df = pd.read_sql_query(QUERY, conn)

# Eliminar valores NaN (nulos) para poder graficar
df = df.dropna()


plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'], bins=5, color='skyblue', edgecolor='black')
plt.title('Histograma de Compensación Convertida (ConvertedComp)')
plt.xlabel('Compensación Convertida (USD)')
plt.ylabel('Frecuencia')
plt.grid(True)

plt.show()


```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
import seaborn as sns

# Ejecutar la consulta SQL para obtener los datos de la columna Age
QUERY = """
SELECT Age
FROM master
"""

df = pd.read_sql_query(QUERY, conn)

# Eliminar valores NaN (nulos) para poder graficar
df = df.dropna()

plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Age', color='lightblue')
plt.title('Diagrama de Caja de la Edad (Age)')
plt.xlabel('Edad (Años)')
plt.grid(True)

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here


# Ejecutar la consulta SQL para obtener los datos de Age y WorkWeekHrs
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

df = pd.read_sql_query(QUERY, conn)

# Eliminar valores NaN (nulos) para poder graficar
df = df.dropna()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='WorkWeekHrs', color='blue')
plt.title('Diagrama de Dispersión: Edad vs Horas de Trabajo Semanales')
plt.xlabel('Edad (Años)')
plt.ylabel('Horas de Trabajo Semanales')
plt.grid(True)

plt.show()

```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

# Ejecutar la consulta SQL para obtener los datos de WorkWeekHrs, CodeRevHrs y Age
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

df = pd.read_sql_query(QUERY, conn)

# Eliminar valores NaN (nulos) para poder graficar
df = df.dropna()


plt.figure(figsize=(10, 6))
bubble_plot = sns.scatterplot(
    data=df, x='WorkWeekHrs', y='CodeRevHrs', size='Age', sizes=(20, 500), hue='Age', palette='cool', legend=False)


plt.title('Gráfico de Burbujas: Horas de Trabajo vs Horas de Revisión de Código (Tamaño por Edad)')
plt.xlabel('Horas de Trabajo Semanales (WorkWeekHrs)')
plt.ylabel('Horas de Revisión de Código (CodeRevHrs)')
plt.grid(True)

plt.show()

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

# Ejecutar la consulta SQL para obtener los datos de la tabla DatabaseDesireNextYear
QUERY = """
SELECT DatabaseDesireNextYear
FROM DatabaseDesireNextYear
"""

df = pd.read_sql_query(QUERY, conn)

# Contar la cantidad de veces que se mencionan las bases de datos y agrupar por las 5 principales
df_counts = df['DatabaseDesireNextYear'].value_counts().nlargest(5)


plt.figure(figsize=(8, 8))
plt.pie(df_counts, labels=df_counts.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title('Top 5 Bases de Datos que los Encuestrados Desean Conocer el Próximo Año')
plt.axis('equal')


plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

# Ejecutar la consulta SQL para obtener los datos de WorkWeekHrs, CodeRevHrs y Age
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age >= 30 AND Age <= 35
"""

df = pd.read_sql_query(QUERY, conn)

# Eliminar valores NaN (nulos) para poder calcular las medianas
df = df.dropna()

# Calcular la mediana de WorkWeekHrs y CodeRevHrs
mediana_workweek = df['WorkWeekHrs'].median()
mediana_coderev = df['CodeRevHrs'].median()

# Crear un dataframe con las medianas
median_data = pd.DataFrame({
    'Horas de Trabajo Semanal (WorkWeekHrs)': [mediana_workweek],
    'Horas de Revisión de Código (CodeRevHrs)': [mediana_coderev]
})

# Dibujar el gráfico apilado
median_data.plot(kind='bar', stacked=True, color=['#66b3ff', '#99ff99'], figsize=(8, 6))

plt.title('Gráfico Apilado: Mediana de Horas de Trabajo y Revisión de Código (Edades 30-35)')
plt.ylabel('Horas')
plt.xticks([])

plt.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

# Ejecutar la consulta SQL para obtener los datos de Age y ConvertedComp
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age >= 45 AND Age <= 60
"""

df = pd.read_sql_query(QUERY, conn)

# Eliminar valores NaN (nulos) en ConvertedComp para calcular la mediana
df = df.dropna(subset=['ConvertedComp'])

# Agrupar por edad y calcular la mediana de ConvertedComp
median_convertedcomp = df.groupby('Age')['ConvertedComp'].median()

# Dibujar el gráfico de la mediana de ConvertedComp por edad
plt.figure(figsize=(10, 6))
plt.plot(median_convertedcomp.index, median_convertedcomp.values, marker='o', color='b', linestyle='-')

plt.title('Mediana de ConvertedComp por Edad (45-60 años)')
plt.xlabel('Edad')
plt.ylabel('Mediana de Compensación Convertida (ConvertedComp)')
plt.grid(True)

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

# Ejecutar la consulta SQL para obtener los datos de la columna MainBranch
QUERY = """
SELECT MainBranch
FROM master
"""

df = pd.read_sql_query(QUERY, conn)

# Contar la frecuencia de cada categoría en MainBranch
mainbranch_counts = df['MainBranch'].value_counts()

# Crear un gráfico de barras horizontales
plt.figure(figsize=(10, 6))
mainbranch_counts.plot(kind='barh', color='skyblue')

plt.title('Distribución de las Categorías de MainBranch')
plt.xlabel('Número de Respuestas')
plt.ylabel('Categorías de MainBranch')

plt.show()
```

```python
# Importar las librerías necesarias

# Conectar a la base de datos
conn = sqlite3.connect("m4_survey_data.sqlite")

# Ejecutar la consulta SQL para obtener los datos de LanguageDesireNextYear
QUERY = """
SELECT LanguageDesireNextYear
FROM LanguageDesireNextYear
"""

df = pd.read_sql_query(QUERY, conn)

# Contar la frecuencia de cada lenguaje
language_counts = df['LanguageDesireNextYear'].value_counts()

# Convertir el conteo en un DataFrame para facilitar la búsqueda
language_counts_df = language_counts.reset_index()
language_counts_df.columns = ['Language', 'Count']

# Buscar el lugar que ocupa Python
python_rank = language_counts_df[language_counts_df['Language'] == 'Python'].index[0] + 1  # Sumar 1 porque el índice empieza en 0

# Mostrar el resultado
print(f"Python ocupa el lugar número {python_rank} entre los idiomas más deseados el próximo año.")
```

```python
# Ejecutar la consulta SQL para obtener los datos de LanguageWorkedWith donde aparece SQL
QUERY = """
SELECT COUNT(*)
FROM LanguageWorkedWith
WHERE LanguageWorkedWith LIKE '%SQL%'
"""

df = pd.read_sql_query(QUERY, conn)

# Mostrar el resultado
sql_count = df.iat[0, 0]
print(f"{sql_count} encuestados indicaron que actualmente trabajan con SQL.")
```

```python

# Ejecutar la consulta SQL para contar los encuestados que trabajan solo con MySQL
QUERY = """
SELECT COUNT(*)
FROM LanguageWorkedWith
WHERE LanguageWorkedWith LIKE '%MySQL%'
"""

df = pd.read_sql_query(QUERY, conn)

# Mostrar el resultado
mysql_count = df.iat[0, 0]
print(f"{mysql_count} encuestados indicaron que trabajan únicamente con MySQL.")
```
